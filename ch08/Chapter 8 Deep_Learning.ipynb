# 딥러닝
## 더 깊게
- 심층 CNN 구조
image.png
- 신경망의 특징
  - 3*3의 작은 필터를 사용한 합성곱 계층
  - 활성화 함수는 ReLU
  - 완전연결 계층 뒤에 드롭아웃 계층 사용
  - Adam을 사용해 최적화
  - 가중치 초깃값은 'He의 초깃값'
- 정확도를 더 높이려면
 - 대부분 CNN을 기초로 한다.
 - 앙상블 학습, 학습률 감소, 데이터 확장 등으로 정확도 향상 시킨다.
  - 데이터 확장: 입력 이미지(훈련 이미지)를 알고리즘을 동원해 '인위적'으로 확장 ex) crop, flip
- 깊게 하는 이유
  - 신경망의 매개변수 수가 줄어든다.
   - 학습 데이터의 양을 줄여 학습을 고속으로 수행할 수 있다.
   - 학습해야 할 문제를 계층적으로 분해할 수 있다.
   - 정보를 계층적으로 전달할 수 있다.

## 딥러닝의 초기 역사
### 이미지넷
 - 100만장이 넘는 이미지를 담고 있는 데이터셋
 - 이미지 인식 기술 대회에서 제대로 분류하는지 겨룬다. -> Top-5 오류 비율 기준
 - 2012년 AlexNet이 오류율을 크게 낮춤
 - 2015년 ResNet이 오류율을 크게 낮추기도 함

### VGG
- 합성곱 계층과 풀링 계층으로 구성되는 '기본적'인 CNN
- 비중 있는 층(합성곱 계층, 완전연결 계층)을 모두 16층(혹은 19층)으로 심화한 게 특징
- 3*3의 작은 필터를 사용한 합성곱 계층을 연속으로 거친다.
- 마지막에는  완전연결 계층을 통과시켜 결과를 출력

### GoogLeNet
- CNN과 기본적으로는 동일
- 세로 방향 깊이뿐 아니라 가로 방향도 깊다는 점이 특징
- 가로 방향에 '폭'이 있으며 이를 인셉션 구조라고 함.
 - 인셉션 구조: 크기가 다른 필터를 여러 개 적용하여 그 결과를 결합한다.

### ResNet
- 지나치게 층을 깊게 하면 성능이 떨어지는 경우가 많아 이를 해결하고자 한다.
- 스킵 연결 도입
  - 입력 데이터를 합성곱 계층을 건너뛰어 출력에 바로 더하는 구조
- 합성곱 계층을 2개 층마다 건너뛰면서 층을 깊게 한다.

## 더 빠르게(딥러닝 고속화)
- GPU 컴퓨팅의 목적: 병렬 수치 연산으로 처리할 수 있으니 압도적인 힘을 다양한 용도로 활용하자
- 딥러닝에 최적화된 라이브러리를 이용하면 더 빨라진다. -> ex) 엠비디아의 GPU 컴퓨팅용 통합 개발환경 CUDA
- 1회 학습에 걸리는 시간을 최대한 단축하기 위한 수평확장(분산 학습)
- 네트워크로 주고받는 데이터의 비트 수를 최소로 만든다. -> 특히 임베디드 분야에서 중요

## 딥러닝의 활용
### 사물 검출
- 이미지 속에 담긴 사물의 위치와 종류(클래스)를 알아내는 기술
- CNN을 이용한 사물 검출 -> R-CNN
 - 사물이 위치한 영역을 어떤 방법으로 찾아내고
 - 추출한 각 영역에 CNN을 적용하여 클래스를 분류한다.
 - Selective Search 기법
- Faster R-CNN : 후보 영역 추출까지 CNN으로 처리한다.

### 분할
- 이미지를 픽셀 수준에서 분류하는 문제
- 픽셀 단위로 객체마다 채색된 지도 데이터를 사용해 학습하고 추론할 때 이미지의 모든 픽셀을 분류
- FCN(Fully Convolutional Network)의 도빕
  - 단 한버의 forward 처리로 모든 픽셀의 클래스를 분류해준다.
  - 공간 볼륨을 유지한 채 마지막 출력까지 처리할 수 있다.
  - 마지막에 공간 크기를 확대하는 처리를 도입하였다.

### 사진 캡션 생성
- 사진을 주면 그 사진을 설명하는 글(사진 캡션)을 자동으로 생성하는 연구
- NIC(Neural Image Cation) 모델: 심층 CNN과 자연어를 다루는 순환 신경망(Recurrent Neural Network, RNN)으로 구성
- 멀티모달 처리: 사진이나 자연어와 같은 여러 종류의 정보를 조합하고 처리한다.

## 딥러닝의 미래
### 이미지 스타일(화풍) 변환
- 두 이미지를 입력해서 새로운 그림을 생성
- 하나는 '콘텐츠 이미지', 다른 하나는 '스타일 이미지'라 부르는데 이 둘을 조합한다.
- 스타일 행렬: 스타이 화풍을 흡수. 스타일 행렬의 오차를 줄이도록 학습한다.

### 이미지 생성
- 아무런 입력 이미지 없이 새로운 이미지 생성
- 대량의 이미지 사용하여 학습하지만 학습 이후에는 아무런 입력 없이 가능하다.
- DCGAN(Deep Convolutional Generative Adversarial Network)
 - 생성자와 식별자로 불리는 2개의 신경망을 이용한다.
 - GAN 기술을 이용해 둘을 겨루도록 한다.

### 자율 주행
- 주위 환경을 올바르게 인식하는 기술이 가장 중요
- SegNet: 주변 환경을 정확하게 인식하는 CNN 기반 신경망

### Deep Q-Network(강화학습)
- 컴퓨터도 시행착오 과정에서 스스로 학습하게 하려는 분야
- 환경에 맞게 행동을 선택하고 그 행동에 의해서 환경이 변한다.
- 이후 보상을 얻고 더 나은 보상을 받는 쪽으로 행동 방향을 정한다.
- Deep Q-Network
  - Q학습에서는 최적 행동 가치 함수로 최적인 행동을 정한다.
  - 이 함수를 딥러닝(CNN)으로 비슷하게 흉내내여 사용한다.
