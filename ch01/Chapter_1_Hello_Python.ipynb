## 1. 클래스 사용하기

```python
class Man:
  def __init__(self, name):
    self.name = name
    print("Initialized!")

  def hello(self):
    print("Hello " + self.name + "!")

  def goodbye(self):
    print("Good-bye " + self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

## 4. Matplotlib로 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y = np.sin(x) #x의 값을 sin으로 구한다.

plt.plot(x, y)
plt.show()
```

### 여러 개 그래프를 그리고 제목, 레이블 표시하기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")

plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
